```python
# soup 객체를 return하는 함수를 정의하기

def get_soup(url):
    HEADERS = {'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
    r = requests.get(url, headers=HEADERS)
    
    if r.status_code == 200:
        print('OK')
        soup = bs(r.text, 'html.parser')
        
    return soup
```

```python
soup = get_soup(url)
print(soup)
```

### 실전! 논문크롤링 - 10편
- 제목, 저자, 발행기관, 연도, 논문지, 상세링크, 초록


```python
# 첫번재 논문 저자
writer = contents[0].find('span', class_='writer').text
print(writer)
```

```python
# 첫번재 논문 발행기관

publisher = contents[-1].find('span', class_='assigned').text
print(publisher)
```

```python
# 첫번째 논문 연도 - p 태그 클래스 etc의 세번째 span 태그

year = contents[0].find('p', class_='etc').find_all('span')[2].text 
print(year)
```

```python
# 학술지
journal = contents[0].find('p', class_='etc').find_all('span')[3].text 
print(journal)
```

```python
#초록
abstracts = contents[0].find('p', class_="preAbstract").text
print(abstracts)
```

```python
# 상세정보 링크

link = 'http://www.riss.kr'+contents[0].find('p', class_='title').find('a')['href']
print(link)
```

```python
# [완성] 10편 논문 수집하고 저장하기


title = []
writer =[]
publisher =[]
year =[]
journal =[]
link = []
abstracts =[]

for cont in contents :
    title.append(cont.find('p', class_='title').text)
    writer.append(cont.find('span', class_='writer').text)
    publisher.append(cont.find('span', class_='assigned').text)
    year.append(cont.find('p', class_='etc').find_all('span')[2].text)
    journal.append(cont.find('p', class_='etc').find_all('span')[3].text)
    link.append('http://www.riss.kr'+ cont.find('p', class_='title').find('a')['href'])
    abstracts.append(cont.find('p', class_='preAbstract').text)
```

```python
# pandas DataFrame으로 저장하기

import pandas as pd

df = pd.DataFrame(
    {'title':title,
    'writer': writer,
    'publisher':publisher,
    'year':year,
    'journal': journal,
    'link': link,
    'abstracts': abstracts}
)
df


```

```python
# soup 객체를 return하는 함수를 정의하기

def get_soup(url):
    
    import requests
    from bs4 import BeautifulSoup as bs
    import pandas as pd
    import time
    
    HEADERS = {'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
    r = requests.get(url, headers=HEADERS)
    
    if r.status_code == 200:
        print('OK')
        soup = bs(r.text, 'html.parser')
        
    return soup
```

```python
soup = get_soup(url)
```

```python
contents = soup.find_all('div', class_= 'cont ml60')
len(contents)
```

```python
# 첫 페이지 크롤링

title = []
writer =[]
publisher =[]
year =[]
journal =[]
link = []
abstracts =[]

for cont in contents :
    title.append(cont.find('p', class_='title').text.strip())
    writer.append(cont.find('span', class_='writer').text)
    publisher.append(cont.find('span', class_='assigned').text)
    year.append(cont.find('p', class_='etc').find_all('span')[2].text)
    journal.append(cont.find('p', class_='etc').find_all('span')[3].text)
    link.append('http://www.riss.kr'+ cont.find('p', class_='title').find('a')['href'])
    
    if cont.find('p', class_='preAbstract'):
        abstracts.append(cont.find('p', class_='preAbstract').text)
        
    else:
        abstracts.append('초록이 없습니다.')
```
